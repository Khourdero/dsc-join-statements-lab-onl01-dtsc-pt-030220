## Connecting to the Database
Import the necessary packages and connect to the database `'data.sqlite'`.

```python
# Your code here
import sqlite3 as sql
import pandas as pd
conn = sql.connect('data.sqlite')
cur = conn.cursor()
```

## Display the names of all the employees in Boston 

Hint: join the employees and offices tables.

```python
# Your code here
cur.execute("""SELECT lastName, firstName, city 
                FROM employees 
                JOIN offices 
                USING(officeCode) 
                WHERE city = 'Boston';""")
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

## Are there any offices that have zero employees?
Hint: Combine the employees and offices tables and use a group by.

```python
# Your code here
cur.execute("""SELECT o.officeCode, o.city, COUNT(e.employeeNumber) AS num_emp
                FROM offices AS o 
                LEFT JOIN employees AS e 
                USING(officeCode)
                GROUP BY officeCode
                HAVING num_emp = 0;""")
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

```python
# Your code here
# What cities are buying the most?

cur.execute("""SELECT c.city, SUM(od.quantityOrdered) AS Num_of_Products
                FROM customers c
                JOIN orders
                USING(customerNumber)
                JOIN orderDetails od
                USING(orderNumber)
                JOIN products p
                USING(productCode)
                GROUP BY c.city
                ORDER BY Num_of_Products DESC;""")
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

```python
# Your code here
# Who bought the most planes?
cur.execute("""SELECT c.customerName, SUM(od.quantityOrdered) AS Products_ordered, p.productLine
                FROM customers c
                JOIN orders
                USING(customerNumber)
                JOIN orderDetails od
                USING(orderNumber)
                JOIN products
                USING(productCode)
                JOIN productLines p
                USING(productLine)
                WHERE p.productLine = 'Planes'
                GROUP BY c.customerName
                ORDER BY Products_ordered DESC;""")
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df.head()
```

## Level Up: Display the names of every individual product that each employee has sold

```python
# Your code here
cur.execute("""SELECT e.firstName, e.lastName, p.productName
                FROM employees e
                JOIN customers c
                ON e.employeeNumber = c.salesRepEmployeeNumber
                JOIN orders
                USING(customerNumber)
                JOIN orderDetails
                USING(orderNumber)
                JOIN products p
                USING(productCode)""")
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```
